# Plotting Logs

```python

# Define curve_names with labels corresponding to each subplot
curve_names = ['Gamma', 'Deep Resistivity','Density', 'Neutron']

fig, axes= plt.subplots(figsize=(10, 10))  # Unpack the tuple to get the figure object

# Plotting Gamma Ray
ax1 = plt.subplot2grid((1, 3), (0, 0), rowspan=1, colspan=1)
ax1.plot(df['GR'], df['DEPTH'], color='green',lw=0.5)  # Ensure the correct order of parameters
ax1.set_xlabel(curve_names[0])  # Set the xlabel using curve_names
ax1.set_xlim(0, 200)

# Plotting Deep Resistivity
ax2 = plt.subplot2grid((1, 3), (0, 1), rowspan=1, colspan=1)
ax2.plot(df['RDEP'], df['DEPTH'], color='red',lw=0.5)  # Ensure the correct order of parameters
ax2.set_xlabel(curve_names[1])  # Set the xlabel using curve_names
ax2.set_xlim(0.2, 2000)
ax2.semilogx()


ax3 = plt.subplot2grid((1, 3), (0, 2), rowspan=1, colspan=1)
ax3.plot(df['DEN'], df['DEPTH'], color='blue',lw=0.5)
ax3.set_xlim(1.95, 2.95)
ax4 = ax3.twiny()

ax4.plot(df['NEU'], df['DEPTH'], color='pink',lw=0.5)
ax4.set_xlim(45, -15)


# Set the y-axis limits and grid
for i,ax in enumerate(fig.axes):  # Iterate over the list of axes
    ax.set_ylim(4700, 3500)  # Ensure the correct orientation for depth (top to bottom)
    ax.xaxis.set_ticks_position('top' )
    ax.xaxis.set_label_position('top')
    ax.set_xlabel(curve_names[i-1])
    if i-1==3:
        ax.spines['top'].set_position(('axes',1.07))
    else:
        ax.grid()

for ax in [ax2, ax3]:
    plt.setp(ax.get_yticklabels(), visible = False)

fig.subplots_adjust(wspace=0.05)

plt.show()
```
